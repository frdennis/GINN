```python
import matplotlib.pyplot as plt
```

```python
from SLoader import SLoader
scaling = 1e-3

loaders = []

stars = ["S1", "S2", "S4", "S8", "S9", "S12", "S13", "S14", "S17", "S18",
        "S19", "S21", "S24", "S31", "S38", "S54", "S55"]

for star in stars:
    SL = SLoader(star=star, scaling=scaling)
    print(star, len(SL))
    loaders.append(SL)


for i, loader in enumerate(loaders):
    fig, ax = plt.subplots(ncols=2, figsize=(10,5))
    fig.suptitle(stars[i])
    ax[0].plot(loader.phi, loader.u, 'rx')
    ax[0].set_ylabel(rf"$1/r$ [({1/scaling} AU)]")
    ax[0].set_xlabel(r"$\phi$ [radians]")
    ax[0].grid()

    ax[1].plot(loader.x, loader.y, 'rx')
    ax[1].set_ylabel(rf"x [{scaling} AU]")
    ax[1].set_xlabel(rf"y [{scaling} AU]")
    ax[1].grid()
    ax[1].axis('equal');
```
